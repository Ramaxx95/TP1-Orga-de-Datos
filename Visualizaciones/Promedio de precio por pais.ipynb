```python
import pandas as pd
import numpy as np
import datetime as datetime
import calendar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/ramil/Documents/UBA/(75.06) Organización de Datos/TPS/TP1/Entrenamieto_ECI_2020.csv')
```

```python
precio = df[['Billing_Country', 'Price', 'Currency']]
precio = precio.loc[precio['Price'] != 'None']
precio = precio.loc[precio['Price'] != 'Other']
precio['Price'] = pd.to_numeric(precio['Price'])
precio
```

```python
precio_usd = precio.loc[precio['Currency'] == 'USD']
precio_eur = precio.loc[precio['Currency'] == 'EUR']
```

```python
promedio_precio_pais_usd = precio_usd.groupby('Billing_Country').agg({'Price':'mean'})
promedio_precio_pais_eur = precio_eur.groupby('Billing_Country').agg({'Price':'mean'})
```

```python
promedio_precio_pais_usd.plot(kind = 'bar', title = 'Promedio del precio por pais emisor de factura(USD)', linewidth = 5, figsize = (15, 8), rot = 80, color = 'lightblue');
ax = plt.gca()
ax.set_ylabel('Promedio en dolares', fontsize = 20)
ax.set_xlabel('Pais', fontsize = 20)

promedio_precio_pais_eur.plot(kind = 'bar', title = 'Promedio del precio por pais emisor de factura(EUR)', linewidth = 5, figsize = (15, 8), rot = 80, color = 'salmon');
ax = plt.gca()
ax.set_ylabel('Promedio en euros', fontsize = 20)
ax.set_xlabel('Pais', fontsize = 20)
```
