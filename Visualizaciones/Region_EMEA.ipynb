```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
oportunidades_cerradas = data_frame[( (data_frame.Stage == 'Closed Won') | (data_frame.Stage == 'Closed Lost')) & (data_frame.Region == 'EMEA') ]
oportunidades_cerradas
```

```python
oportunidades_ganadas_EMEA = data_frame[(data_frame.Stage == 'Closed Won') & (data_frame.Region == 'EMEA')]
oportunidades_ganadas_EMEA
```

```python
oportunidades_perdidas_EMEA = data_frame[(data_frame.Stage == 'Closed Lost') & (data_frame.Region == 'EMEA')]
oportunidades_perdidas_EMEA
```

```python
oportunidades_ganadas_middle_EMEA_df =  oportunidades_ganadas_EMEA.Billing_Country.value_counts()
oportunidades_ganadas_middle_EMEA_df = oportunidades_ganadas_middle_EMEA_df.to_frame()
oportunidades_ganadas_middle_EMEA_df.head(10)
```

```python
oportunidades_ganadas_middle_EMEA_df.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos cerrados ganados de EMEA', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('oportunidades cerradas ganadas');
ax.set_xlabel('Paises')
```

```python
oportunidades_perdidas_middle_EMEA_df =  oportunidades_perdidas_EMEA.Billing_Country.value_counts()
oportunidades_perdidas_middle_EMEA_df = oportunidades_perdidas_middle_EMEA_df.to_frame()
oportunidades_perdidas_middle_EMEA_df.head(10)
```

```python
oportunidades_perdidas_middle_EMEA_df.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos cerrados perdidas de EMEA', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('oportunidades cerradas perdidas');
ax.set_xlabel('Paises')
```

```python
anio_2017 = oportunidades_cerradas[ oportunidades_cerradas.Opportunity_Created_Date.dt.year == 2017]
anio_2017.Stage.value_counts().to_frame().plot.pie(title = 'Distribucion de casos en 2017 en la Region de EMEA', y = 'Stage', figsize = (11,6))
plt.show()
```

```python
paises_cerrados_2017 = anio_2017.Territory.value_counts().to_frame( name = 'Oportunidades cerradas')
paises_cerrados_2017.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos cerrados en 2017', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('Oportunidades cerradas');
ax.set_xlabel('Paises')
plt.show()
```

```python
anio_2017_ganados = anio_2017[anio_2017.Stage == 'Closed Won']
grafico_2017_ganados = anio_2017_ganados.Territory.value_counts().to_frame( name = 'Paises con oportunidades ganadas')
grafico_2017_ganados.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos ganados en 2017', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('Oportunidades cerradas ganadas');
ax.set_xlabel('Paises')
plt.show()
```
