```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
oportunidades_cerradas = data_frame[( (data_frame.Stage == 'Closed Won') | (data_frame.Stage == 'Closed Lost')) & (data_frame.Region == 'EMEA') ]
oportunidades_cerradas
```

```python
oportunidades_ganadas_EMEA = data_frame[(data_frame.Stage == 'Closed Won') & (data_frame.Region == 'EMEA')]
oportunidades_ganadas_EMEA
```

```python
oportunidades_perdidas_EMEA = data_frame[(data_frame.Stage == 'Closed Lost') & (data_frame.Region == 'EMEA')]
oportunidades_perdidas_EMEA
```

```python
anios_oportunidades_cerradas = oportunidades_cerradas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Cantidad oportunidades cerradas').sort_index()
anios_oportunidades_cerradas
```

```python
anios_oportunidades_ganadas = oportunidades_ganadas_EMEA.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Cantidad oportunidades ganadas').sort_index()
anios_oportunidades_ganadas
```

```python
anios_oportunidades_perdidas = oportunidades_perdidas_EMEA.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Cantidad oportunidades perdidas').sort_index()
anios_oportunidades_perdidas
```

```python
lista_anios = [anios_oportunidades_ganadas, anios_oportunidades_cerradas, anios_oportunidades_perdidas]
anios_totales =  pd.concat(lista_anios, join = 'outer', axis = 1).fillna(0)
anios_totales
```

```python
anios_totales.plot.bar( title = 'Oportunidades totales, ganadas y perdidas por anio creadas' , rot = 65, figsize = ((20,10)), color = ['green', 'blue', 'red'])
ax=plt.gca()
ax.set_ylabel('Cantidad de casos cerrados');
ax.set_xlabel('Anios')
plt.show()
```

```python
oportunidades_ganadas_middle_EMEA_df =  oportunidades_ganadas_EMEA.Billing_Country.value_counts()
oportunidades_ganadas_middle_EMEA_df = oportunidades_ganadas_middle_EMEA_df.to_frame( name = 'Oportunidades Ganadas').head(10)
oportunidades_ganadas_middle_EMEA_df
```

```python
oportunidades_ganadas_middle_EMEA_df.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Top 10 de paises con casos ganados a lo largo de los anios', color = 'green');
ax=plt.gca()
ax.set_ylabel('oportunidades cerradas ganadas');
ax.set_xlabel('Paises')
plt.show()
```

```python
oportunidades_perdidas_middle_EMEA_df =  oportunidades_perdidas_EMEA.Billing_Country.value_counts()
oportunidades_perdidas_middle_EMEA_df = oportunidades_perdidas_middle_EMEA_df.to_frame( name =  'Oportunidades perdidas').head(10)
oportunidades_perdidas_middle_EMEA_df
```

```python
oportunidades_perdidas_middle_EMEA_df.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Top 10 paises con casos cerrados perdidas de EMEA', color = 'red');
ax=plt.gca()
ax.set_ylabel('oportunidades cerradas perdidas');
ax.set_xlabel('Paises')
plt.show()
```

```python
anio_2017 = oportunidades_cerradas[ oportunidades_cerradas.Opportunity_Created_Date.dt.year == 2017]
anio_2017.Stage.value_counts().to_frame().plot.pie(title = 'Distribucion de casos en 2017 en la Region de EMEA', y = 'Stage', figsize = (11,6), autopct='%1.0f%%')
plt.show()
```

```python
paises_cerrados_2017 = anio_2017.Billing_Country.value_counts().to_frame( name = 'Oportunidades cerradas').nlargest(10, 'Oportunidades cerradas')
paises_cerrados_2017
```

```python
paises_cerrados_2017.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos cerrados creados en 2017', color = 'blue');
ax=plt.gca()
ax.set_ylabel('Oportunidades cerradas');
ax.set_xlabel('Paises')
plt.show()
```

```python
anio_2017_ganados = anio_2017[anio_2017.Stage == 'Closed Won']
grafico_2017_ganados = anio_2017_ganados.Billing_Country.value_counts().to_frame( name = 'Paises con oportunidades ganadas').nlargest(10, 'Paises con oportunidades ganadas')
grafico_2017_ganados.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos ganados creados en 2017', color = 'green');
ax=plt.gca()
ax.set_ylabel('Oportunidades cerradas ganadas');
ax.set_xlabel('Paises')
plt.show()
```

```python
anio_2017_perdidos = anio_2017[anio_2017.Stage == 'Closed Lost']
grafico_2017_perdidos_pais = anio_2017_perdidos.Territory.value_counts().to_frame( name = 'Paises con oportunidades ganadas').nlargest(10, 'Paises con oportunidades ganadas')
grafico_2017_perdidos_pais.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos perdidos creados en 2017', color = 'red');
ax=plt.gca()
ax.set_ylabel('Oportunidades cerradas perdidas');
ax.set_xlabel('Paises')
plt.show()
```

```python
anio_2017_vendedores_totales = anio_2017.Account_Owner.value_counts().to_frame(name = 'Oportunidades Cerradas')
anio_2017_vendedores_totales
```

```python
anio_2017_vendedores_ganados = anio_2017_ganados.Account_Owner.value_counts().to_frame( name = 'Oportunidades Ganadas')
anio_2017_vendedores_ganados
```

```python
anio_2017_vendedores_perdidos = anio_2017_perdidos.Account_Owner.value_counts().to_frame( name = 'Oportunidades Perdidas')
anio_2017_vendedores_perdidos
```

```python
# Para un grafico de barras agrupado lo que vamos a hacer es a esos 3 value counts lo vamos a llevar a un data frame y ponerlos todos en un mismo lugar
frames = [anio_2017_vendedores_ganados, anio_2017_vendedores_totales, anio_2017_vendedores_perdidos]
grafico =  pd.concat(frames, join = 'outer', axis = 1).fillna(0).nlargest(5, 'Oportunidades Ganadas')
grafico
```

```python
grafico.plot.barh( title = 'Top 5 vendedores en el 2017 por cantidad de oportunidades ganadas' , rot = 65, figsize = ((20,10)), color = ['green', 'blue', 'red'])
plt.show()
```
