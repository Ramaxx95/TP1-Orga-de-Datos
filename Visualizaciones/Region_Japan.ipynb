```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
japon_cerrados_df =  data_frame[ ( ((data_frame.Stage == 'Closed Won') |  (data_frame.Stage == 'Closed Lost') )& (data_frame.Region == 'Japan'))]
japon_cerrados_df
```

```python
japon_cerrados_ganadas_df = data_frame[(data_frame.Stage == 'Closed Won') & (data_frame.Region == 'Japan')]
japon_cerrados_ganadas_df
```

```python
japon_cerrados_perdidas_df = data_frame[(data_frame.Stage == 'Closed Lost') & (data_frame.Region == 'Japan')]
japon_cerrados_perdidas_df
```

```python
anios_totales = japon_cerrados_df.Opportunity_Created_Date.dt.year.value_counts().sort_index().to_frame(name = 'Cantidad de oportunidades creadas')
anios_totales
```

```python
anios_ganados = japon_cerrados_ganadas_df.Opportunity_Created_Date.dt.year.value_counts().to_frame(name = 'Cantidad oportunidadas creadas que se concretaron').sort_index()
anios_ganados
```

```python
anios_perdidos = japon_cerrados_perdidas_df.Opportunity_Created_Date.dt.year.value_counts().to_frame(name = 'Cantidad oportunidades creadas no concretadas').sort_index()
anios_perdidos
```

```python
lista_anios = [anios_ganados, anios_totales, anios_perdidos]
casos_totales_por_anio = pd.concat(lista_anios, join = 'outer', axis = 1).fillna(0)
casos_totales_por_anio
```

```python
casos_totales_por_anio.plot.bar( title = 'Casos cerrados ganados, totales y perdidos por anio de creacion', rot = 0 ,figsize = ((20,10)), color = ['green', 'blue', 'red'])
ax=plt.gca()
ax.set_ylabel('Cantidad de casos abiertos');
ax.set_xlabel('Anio')
plt.show()
```

```python
vendedores_totales = japon_cerrados_df.Account_Owner.value_counts().to_frame(name = 'Cantidad de casos cerrados totales')
vendedores_totales
```

```python
vendedores_ganados = japon_cerrados_ganadas_df.Account_Owner.value_counts().to_frame(name = 'Oportunidades ganadas')
vendedores_ganados
```

```python
vendedores_perdidos = japon_cerrados_perdidas_df.Account_Owner.value_counts().to_frame( name = 'Oportunidades perdidas')
vendedores_perdidos
```

```python
lista_vendedores = [vendedores_ganados, vendedores_totales, vendedores_perdidos]
top_5_vendedores = pd.concat(lista_vendedores, join = 'outer', axis = 1).fillna(0).nlargest(5, 'Oportunidades ganadas')
top_5_vendedores
```

```python
top_5_vendedores.plot.barh( title = 'Top 5 de vendedores en Japon a lo largo de los anios', rot = 0 ,figsize = ((20,10)), color = ['green', 'blue', 'red'])
ax=plt.gca()
ax.set_ylabel('Vendedores');
ax.set_xlabel('Cantidad de casos')
plt.show()
```

```python
grafico_casos_perdidos_creados_2015_df = casos_perdidos_japon_2015_df.Opportunity_Created_Date.dt.month.value_counts().to_frame().sort_index()
grafico_casos_perdidos_creados_2015_df.columns=['Cantidad de casos abiertos en la fecha']
grafico_casos_perdidos_creados_2015_df
```

```python
grafico_casos_perdidos_creados_2015_df.plot(kind = 'bar', figsize =(16,8), rot =65, title = 'Cantidad de casos perdidos abiertos 2015, por mes', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('Cantidad de casos abiertos');
ax.set_xlabel('Fecha')
```

```python
japon_solo_anios_y_oportunidades_cerradas = japon_cerrados_df[['Stage','Opportunity_Created_Date']]
anio_cuentas_creadas = japon_solo_anios_y_oportunidades_cerradas.Opportunity_Created_Date.dt.year
japon_solo_anios_y_oportunidades_cerradas.loc[:,'Opportunity_Created_Date'] = anio_cuentas_creadas
#japon_solo_anios_y_oportunidades_cerradas.columns = ['Stage', 'Anio_de_oportunidad_creada']
#japon_solo_anios_y_oportunidades_cerradas.set_index(japon_solo_anios_y_oportunidades_cerradas.Stage, drop = True, inplace = True)
japon_solo_anios_y_oportunidades_cerradas
```

```python
casos_2015 = japon_solo_anios_y_oportunidades_cerradas[japon_solo_anios_y_oportunidades_cerradas.Opportunity_Created_Date == 2015]
casos_2015 = casos_2015.Stage.value_counts().to_frame(name = 'Anio_2015')
casos_2016 = japon_solo_anios_y_oportunidades_cerradas[japon_solo_anios_y_oportunidades_cerradas.Opportunity_Created_Date == 2016]
casos_2016 = casos_2016.Stage.value_counts().to_frame(name = 'Anio_2016')
casos_2017 = japon_solo_anios_y_oportunidades_cerradas[japon_solo_anios_y_oportunidades_cerradas.Opportunity_Created_Date == 2017]
casos_2017 = casos_2017.Stage.value_counts().to_frame(name = 'Anio_2017')
casos_2018 = japon_solo_anios_y_oportunidades_cerradas[japon_solo_anios_y_oportunidades_cerradas.Opportunity_Created_Date == 2018]
casos_2018 = casos_2018.Stage.value_counts().to_frame(name = 'Anio_2018')
lista_a_unir = [ casos_2015,casos_2016,casos_2017,casos_2018]
casos_cerrados_por_anios =  pd.concat(lista_a_unir, join = 'outer', axis = 1).fillna(0)
#casos_2015[['Anio_2016','Anio_2017']] = [casos_2016.Anio_2016, casos_2017.Anio_2017]
#casos_2015['Anio_2018']= casos_2018.Anio_2018
casos_cerrados_por_anios 
```

```python
casos_cerrados_por_anios.plot.pie( subplots = True, figsize=(25, 10), autopct='%1.0f%%', title = 'Evolucion de la distribucion de los casos en la region de Japon')
plt.show()
```
