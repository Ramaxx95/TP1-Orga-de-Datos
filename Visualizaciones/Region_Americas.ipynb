```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
oportunidades_cerradas = data_frame[ ((data_frame.Stage == 'Closed Lost') | (data_frame.Stage == 'Closed Won')) & (data_frame.Region == 'Americas')]
oportunidades_cerradas
```

```python
oportunidades_perdidas_Americas = data_frame[(data_frame.Stage == 'Closed Lost') & (data_frame.Region == 'Americas')]
oportunidades_perdidas_Americas
```

```python
oportunidades_ganadas_Americas = data_frame[(data_frame.Stage == 'Closed Won') & (data_frame.Region == 'Americas')]
oportunidades_ganadas_Americas
```

```python
ganadas_por_pais_Americas_df = oportunidades_ganadas_Americas.Billing_Country.value_counts().to_frame(name = 'Cantidad oportunidades ganadas por pais')
ganadas_por_pais_Americas_df
```

```python
perdidos_por_pais_Americas_df = oportunidades_perdidas_Americas.Billing_Country.value_counts().to_frame(name = 'Cantidad oportunidades perdidas')
perdidos_por_pais_Americas_df
```

```python
paises_totales = oportunidades_cerradas.Billing_Country.value_counts().to_frame(name = 'Oportunidades Cerradas')
paises_totales
```

```python
lista_paises = [ganadas_por_pais_Americas_df, paises_totales, perdidos_por_pais_Americas_df]
top3_Americas = pd.concat(lista_paises, join = 'outer', axis = 1).fillna(0).nlargest(3, 'Cantidad oportunidades ganadas por pais')
top3_Americas
```

```python
top3_Americas.plot.barh( title = 'Top 3 paises Americas' , rot = 0, figsize = ((20,10)), color = ['green','blue','red'])
ax=plt.gca()
ax.set_ylabel('Paises');
ax.set_xlabel('Cantidad oportunidades')
plt.show()
```

```python
oportunidades_totales_por_anio = oportunidades_cerradas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Oportunidades Cerradas').sort_index()
oportunidades_totales_por_anio
```

```python
oportunidades_ganadas_por_anio = oportunidades_ganadas_Americas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Oportunidades Ganadas').sort_index()
oportunidades_ganadas_por_anio
```

```python
oportunidades_perdidas_por_anio = oportunidades_perdidas_Americas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Oportunidades Perdidas').sort_index()
oportunidades_perdidas_por_anio
```

```python
frames = [oportunidades_ganadas_por_anio, oportunidades_totales_por_anio, oportunidades_perdidas_por_anio]
grafico =  pd.concat(frames, join = 'outer', axis = 1).fillna(0)
grafico
```

```python
grafico.plot.bar( title = 'Cantidad de oportunidades creadas y cerradas por año' , rot = 0, figsize = ((20,10)), color = ['green','blue','red'])
ax=plt.gca()
ax.set_ylabel(' Cantidad oportunidades');
ax.set_xlabel('Años')
plt.show()
```

```python
total_vendedores = oportunidades_cerradas[oportunidades_cerradas.Opportunity_Created_Date.dt.year == 2018].Account_Owner.value_counts().to_frame( name = 'Total oportunidades creadas')
total_vendedores
```

```python
ganadas_vendedores = oportunidades_ganadas_Americas[oportunidades_ganadas_Americas.Opportunity_Created_Date.dt.year == 2018].Account_Owner.value_counts().to_frame( name = 'Oportunidades ganadas')
ganadas_vendedores
```

```python
perdidas_vendedores = oportunidades_perdidas_Americas[oportunidades_perdidas_Americas.Opportunity_Created_Date.dt.year == 2018].Account_Owner.value_counts().to_frame( name = 'Oportunidades perdidas')
perdidas_vendedores
```

```python
lista_vendedores = [ ganadas_vendedores, total_vendedores, perdidas_vendedores]
top_5_vendedores = pd.concat(lista_vendedores, join = 'outer', axis = 1).fillna(0).nlargest(5, 'Oportunidades ganadas')
top_5_vendedores
```

```python
top_5_vendedores.plot.barh( title = 'Mejores 5 vendedores con oportunidades ganadas de las Americas en el 2018' , rot = 0, figsize = ((20,10)), color = ['green','blue','red'])
ax=plt.gca()
ax.set_ylabel('Vendedores');
ax.set_xlabel('Cantidad de casos')
plt.show()
```
