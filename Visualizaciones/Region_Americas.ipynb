```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
oportunidades_cerradas = data_frame[ ((data_frame.Stage == 'Closed Lost') | (data_frame.Stage == 'Closed Won')) & (data_frame.Region == 'Americas')]
oportunidades_cerradas
```

```python
oportunidades_perdidas_Americas = data_frame[(data_frame.Stage == 'Closed Lost') & (data_frame.Region == 'Americas')]
oportunidades_perdidas_Americas
```

```python
oportunidades_ganadas_Americas = data_frame[(data_frame.Stage == 'Closed Won') & (data_frame.Region == 'Americas')]
oportunidades_ganadas_Americas
```

```python
ganadas_por_pais_Americas_df = oportunidades_ganadas_Americas.Billing_Country.value_counts().to_frame()
ganadas_por_pais_Americas_df.columns = ['Cantidad oportunidades ganadas']
ganadas_por_pais_Americas_df
```

```python
print('Posiblemente los graficos de barra no sean lo mejor, puede ser un histograma u otra cosa')
```

```python
ganadas_por_pais_Americas_df.plot(kind= 'bar',figsize =(16,8), rot =0, title = 'Paises con casos ganados en las Americas', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel(' Cantidad oportunidades ganadas');
ax.set_xlabel('Paises')
plt.show()
```

```python
perdidos_por_pais_Americas_df = oportunidades_perdidas_Americas.Billing_Country.value_counts().to_frame()
perdidos_por_pais_Americas_df.columns = ['Cantidad oportunidades perdidas']
perdidos_por_pais_Americas_df
```

```python
print('Hay que ver de unir todos los paises y remarcar cual seria el total de oportunidades perdidas y oportunidades ganadas, eso o como habia hecho mica, agarrar el top 10/20')
```

```python
perdidos_por_pais_Americas_df.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos perdidos en las Americas', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel(' Cantidad oportunidades perdidas');
ax.set_xlabel('Paises')
plt.show()
```

```python
oportunidades_totales_por_anio = oportunidades_cerradas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Oportunidades Cerradas').sort_index()
oportunidades_totales_por_anio
```

```python
oportunidades_ganadas_por_anio = oportunidades_ganadas_Americas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Oportunidades Ganadas').sort_index()
oportunidades_ganadas_por_anio
```

```python
oportunidades_perdidas_por_anio = oportunidades_perdidas_Americas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = 'Oportunidades Perdidas').sort_index()
oportunidades_perdidas_por_anio
```

```python
frames = [oportunidades_ganadas_por_anio, oportunidades_totales_por_anio, oportunidades_perdidas_por_anio]
grafico =  pd.concat(frames, join = 'outer', axis = 1).fillna(0)
grafico
```

```python
grafico.plot.bar( title = 'Cantidad de oportunidades creadas y cerradas por anio' , rot = 0, figsize = ((20,10)))
plt.show()
```

```python
total_vendedores = oportunidades_cerradas.Account_Owner.value_counts().to_frame( name = 'Total oportunidades creadas')
total_vendedores
```

```python
ganadas_vendedores = oportunidades_ganadas_Americas.Account_Owner.value_counts().to_frame( name = 'Oportunidades ganadas')
ganadas_vendedores
```

```python
perdidas_vendedores = oportunidades_perdidas_Americas.Account_Owner.value_counts().to_frame( name = 'Oportunidades perdidas')
perdidas_vendedores
```

```python
lista_vendedores = [ ganadas_vendedores, total_vendedores, perdidas_vendedores]
top_5_vendedores = pd.concat(lista_vendedores, join = 'outer', axis = 1).fillna(0).sort_values(by = 'Oportunidades ganadas', ascending = False).head(5)
top_5_vendedores
```

```python
top_5_vendedores.plot.barh( title = 'Mejores 5 vendedores con oportunidades de las Americas a lo largo de los anios' , rot = 0, figsize = ((20,10)))
plt.show()
```
