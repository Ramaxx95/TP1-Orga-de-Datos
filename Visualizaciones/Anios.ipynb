```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
oportunidades_cerradas = data_frame[(data_frame.Stage == 'Closed Lost') | (data_frame.Stage == 'Closed Won') ]
```

```python
cantidad_cerradas_por_anio = oportunidades_cerradas.Opportunity_Created_Date.dt.year.value_counts().to_frame()
cantidad_cerradas_por_anio.columns = ['Cantidad de oportunidades creadas por año']
cantidad_cerradas_por_anio.sort_index(inplace = True)
cantidad_cerradas_por_anio
```

```python
cantidad_cerradas_por_anio.plot(kind= 'bar',figsize =(16,8), rot =0, title = 'Cantiddad de oportunidades abiertas por anio', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel(' Cantidad oportunidades');
ax.set_xlabel('Años')
```

```python
cantidad_cerradas_por_anio.plot.pie(y= 'Cantidad de oportunidades creadas por año', figsize=(11, 6))
```

```python
oportunidades_ganadas = data_frame[(data_frame.Stage == 'Closed Won')]
oportunidades_ganadas
```

```python
oportunidades_ganadas_por_anio_creacion = oportunidades_ganadas.Opportunity_Created_Date.dt.year.value_counts().sort_index().to_frame( name = 'Cantidad oportunidades ganadas por anio de creacion')
oportunidades_ganadas_por_anio_creacion
```

```python
# Hay que pasarlo a un pie o a ora cosa, algo distinto a un grafico de barras
oportunidades_ganadas_por_anio_creacion.plot(kind= 'bar',figsize =(16,8), rot =0, title = 'Cantiddad de oportunidades abiertas ganadas por anio', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel(' Cantidad oportunidades ganadas abiertas');
ax.set_xlabel('Años')
```

```python
oportunidades_perdidas = data_frame[(data_frame.Stage == 'Closed Lost')]
oportunidades_perdidas.Stage.value_counts()
```

```python
oportunidades_perdidas_por_anio_creacion = oportunidades_perdidas.Opportunity_Created_Date.dt.year.value_counts().sort_index().to_frame( name = 'Cantidad oportunidades perdidas por anio de creacion')
oportunidades_perdidas_por_anio_creacion.drop(index=2013,inplace = True)
oportunidades_perdidas_por_anio_creacion
```

```python
# Hay que pasarlo a un pie o a ora cosa, algo distinto a un grafico de barras
oportunidades_perdidas_por_anio_creacion.plot(kind= 'bar',figsize =(16,8), rot =0, title = 'Cantiddad de oportunidades perdidas ganadas por anio', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel(' Cantidad oportunidades perdidas abiertas');
ax.set_xlabel('Años')
```

```python
#define x and y limits
y = range(0, len(oportunidades_perdidas_por_anio_creacion))
x_perdidos = oportunidades_perdidas_por_anio_creacion['Cantidad oportunidades perdidas por anio de creacion']
x_ganados = oportunidades_ganadas_por_anio_creacion['Cantidad oportunidades ganadas por anio de creacion']

#define plot parameters
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(9, 6))

#specify background color and plot title
fig.patch.set_facecolor('xkcd:light grey')
plt.figtext(.5,.9,"Population Pyramid ", fontsize=15, ha='center')

#define male and female bars
axes[0].barh(y, x_perdidos, align='center', color='royalblue')
axes[0].set(title='Perdidos')
axes[1].barh(y, x_ganados, align='center', color='lightpink')
axes[1].set(title='Ganados')

#adjust grid parameters and specify labels for y-axis
axes[1].grid()
axes[0].set(yticks=y, yticklabels=)
axes[0].invert_xaxis()
axes[0].grid()

plt.show()
```
