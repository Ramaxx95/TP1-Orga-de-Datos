```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('Entrenamieto_ECI_2020.csv')
data_frame.Account_Created_Date = pd.to_datetime(data_frame.Account_Created_Date)
data_frame.Actual_Delivery_Date = pd.to_datetime(data_frame.Actual_Delivery_Date)
data_frame.Opportunity_Created_Date = pd.to_datetime(data_frame.Opportunity_Created_Date)
data_frame.Quote_Expiry_Date = pd.to_datetime(data_frame.Quote_Expiry_Date)
data_frame.Last_Modified_Date = pd.to_datetime(data_frame.Last_Modified_Date)
data_frame.Planned_Delivery_Start_Date = pd.to_datetime(data_frame.Planned_Delivery_Start_Date)
data_frame
```

```python
oportunidades_cerradas = data_frame[ ((data_frame.Stage == 'Closed Lost') | (data_frame.Stage == 'Closed Won')) & ( data_frame.Region ==  'Middle East') ]
oportunidades_cerradas
```

```python
oportunidades_ganadas_middle_east = data_frame[(data_frame.Stage == 'Closed Won') & (data_frame.Region == 'Middle East')]
oportunidades_ganadas_middle_east
```

```python
oportunidades_perdidas_middle_east = data_frame[(data_frame.Stage == 'Closed Lost') & (data_frame.Region == 'Middle East')]
oportunidades_perdidas_middle_east
```

```python
oportunidades_ganadas_middle_east_df_paises_df =  oportunidades_ganadas_middle_east.Billing_Country.value_counts()
oportunidades_ganadas_middle_east_df_paises_df = oportunidades_ganadas_middle_east_df_paises_df.to_frame()
oportunidades_ganadas_middle_east_df_paises_df
```

```python
oportunidades_ganadas_middle_east_df_paises_df.plot(kind= 'bar',figsize =(16,8), rot =0, title = 'Paises con casos cerrados ganados del middle east', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('oportunidades cerradas perdidas');
ax.set_xlabel('Paises')
```

```python
oportunidades_perdidas_middle_east_df_paises_df =  oportunidades_perdidas_middle_east.Billing_Country.value_counts()
oportunidades_perdidas_middle_east_df_paises_df = oportunidades_perdidas_middle_east_df_paises_df.to_frame()
oportunidades_perdidas_middle_east_df_paises_df
```

```python
oportunidades_perdidas_middle_east_df_paises_df.plot(kind= 'bar',figsize =(16,8), rot =65, title = 'Paises con casos cerrados perdidos del middle east', color = (0.7, 0.2, 0.4, 0.6));
ax=plt.gca()
ax.set_ylabel('oportunidades cerradas perdidas');
ax.set_xlabel('Paises')
```

```python
anios_totales = oportunidades_cerradas.Opportunity_Created_Date.dt.year.value_counts().to_frame( name = ' Cantidad de oportunidades creadas').sort_index()
anios_totales
```

```python
anios_ganados = oportunidades_ganadas_middle_east.Opportunity_Created_Date.dt.year.value_counts().to_frame(name = 'Cantidad de oportunidades ganadas creadas').sort_index()
anios_ganados
```

```python
anios_perdidos = oportunidades_perdidas_middle_east.Opportunity_Created_Date.dt.year.value_counts().to_frame(name = 'Cantidad de oportunidades perdidas').sort_index()
anios_perdidos
```

```python
lista_anios = [anios_ganados, anios_totales, anios_perdidos]
grafico_anios = pd.concat(lista_anios, join = 'outer', axis = 1).fillna(0)
grafico_anios
```

```python
grafico_anios.plot.bar( title = 'Cantidad oportunidades totales, ganadas y perdidas por anio' , rot = 0, figsize = ((20,10)))
plt.show()
```

```python
vendedores_totales = oportunidades_cerradas.Account_Owner.value_counts().to_frame(name = 'Oportunidades creadas po vendedor')
vendedores_totales
```

```python
vendedores_ganados = oportunidades_ganadas_middle_east.Account_Owner.value_counts().to_frame(name = 'Oportunidades ganadas por vendedor')
vendedores_ganados
```

```python
vendedores_perdidos = oportunidades_perdidas_middle_east.Account_Owner.value_counts().to_frame(name = 'Oportunidades perdidas por vendedor')
vendedores_perdidos
```

```python
ventas_totales_vendedores = [vendedores_ganados, vendedores_totales, vendedores_perdidos]
grafico_vendedores =  pd.concat(ventas_totales_vendedores, join = 'outer', axis = 1).fillna(0).nlargest(5, 'Oportunidades ganadas por vendedor')
grafico_vendedores
```

```python
grafico_vendedores.plot.barh( title = 'Top 5 vendedores ventas totales por cantidad de oportunidades ganadas' , rot = 65, figsize = ((20,10)))
plt.show()
```
